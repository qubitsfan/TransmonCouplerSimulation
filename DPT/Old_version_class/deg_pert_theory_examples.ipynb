```python
## Other tests
## Operators of Lindgren and Morrison
def Omega(n):
    """The Omega of Lindgren is actually the same as the L of Soliverez"""
    P0 = Operator(OperatorString([0], 1))
    R = Operator(OperatorString([1], 1))
    if n == 0:
        return P0
    if n == 1:
        return R.Vdot(P0)
    else:
        result = R.Vdot((Omega(n-1)))
        for m in range(1, n):
            result -= R*Omega(n-m)*P0.Vdot(Omega(m-1))*P0
        return result
    
def Heff_Lindgren(n):
    P0 = Operator(OperatorString([0], 1))
    return P0.Vdot(Omega(n-1))*P0

# n = 4
# print(Omega(n) - L(n))

print("Heff Lindgren")
print(Heff_Lindgren(1))
print(Heff_Lindgren(2))
print(Heff_Lindgren(3))
print(Heff_Lindgren(4))
print(Heff_Lindgren(5))
print(Heff_Lindgren(6))
# print(Heff_Lindgren(6))
# print(Heff_Lindgren(7))
# print(Heff_Lindgren(9))
```
